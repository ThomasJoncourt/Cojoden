```python
import pandas as pd
import mysql.connector
```

```python
table_oeuvre_auteur = pd.read_csv("table_oeuvre_auteur.csv")
table_musee = pd.read_csv("table_musee.csv")
table_Oeuvre_Domaine = pd.read_csv("table_Oeuvre_Domaine.csv")
table_oeuvres = pd.read_csv("table_oeuvres.csv")
```

```python
cnx = mysql.connector.connect(user='root',password = "root", database='joconde2')
```

```python
cursor = cnx.cursor()
for row in table_musee.index:
    add_musee = ("INSERT INTO musees"
                    "(Id, Nom, Region, Departement, Ville, Latitude, Longitude) "
                    "VALUES (%s, %s, %s, %s, %s, %s, %s)")
    data_musee = (table_musee.Identifiant[row], table_musee['Nom officiel'][row], table_musee.Région[row], table_musee.Département[row], table_musee.Ville[row], table_musee.Latitude[row],table_musee.Longitude[row])
    cursor.execute(add_musee, data_musee)
    cnx.commit()
cursor.close()
cnx.close()
```

```python
cnx = mysql.connector.connect(user='root',password = "root", database='joconde2')
cursor = cnx.cursor()
for row in table_oeuvres.index:
    try:
        add = ("INSERT INTO oeuvres"
                        "(Id, Titre, Période_Création, IdMusee) "
                        "VALUES (%s, %s, %s, %s)")
        data = (table_oeuvres['ID-notice'][row], table_oeuvres.Titre[row], table_oeuvres["Période de création"][row], table_oeuvres["Identifiant Museofile"][row])
        cursor.execute(add, data)
        cnx.commit()
    except:
        pass
cursor.close()
cnx.close()
```
